# This data from 1985 Ward's Automotive Yearbook For Used Cars

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)    #Since the df has no header
df
```

```python
#Adding a header

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

df.columns = headers
df
```

## Data Cleaning

### Change the variables' types to the correct ones

We need to change the type of the following variables from object to int64 or float64:
* normalized-losses
* curb-weight
* bore	
* stroke  
* horsepower 
* peak-rpm 
* price

```python
cols_to_convert = ["normalized-losses", "curb-weight", "bore", "stroke", "horsepower", "peak-rpm", "price"]
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors="coerce")   
```

### Identify missing values

```python
import numpy as np

#we need to replace the "?" symbol with NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Count missing values in each column

df.isnull().sum()
```

#### we found out that the following columns have:
1. "normalized-losses": 41 missing data
2. "num-of-doors": 2 missing data
3. "bore": 4 missing data
4. "stroke" : 4 missing data
5. "horsepower": 2 missing data
6. "peak-rpm": 2 missing data
7. "price": 4 missing data

### Handle missing values

```python
#Fill missing numerical values with median
for col in ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm"]:
    df[col].fillna(df[col].mean(), inplace=True)

#Fill categorical values with mode
df["num-of-doors"].fillna(df["num-of-doors"].mode()[0], inplace=True)

#Drop missing values along the column "price".
df.dropna(subset=["price"], axis=0, inplace = True)
```

## Data Standardization

```python
#transforming meter/gallon to liter/100km in columns "city-mpg" & "highway-mpg" and then renaming them
#L/100km = 235 / mpg

cols_to_transform = ["city-mpg", "highway-mpg"]
df[cols_to_transform] = 235 / df[cols_to_transform]

df.rename(columns = {"city-mpg" : "city-L/100km", "highway-mpg" : "highway-L/100km"}, inplace = True)
df
```

## Data Normalization

```python
#appling the simple feature scalling method on the "length", "width" and "height" columns 
#replace the original value by (original value)/(maximum value) to make their values range from 0 to 1

cols_to_normalize = ["length", "width", "height"]
df[cols_to_normalize] = df[cols_to_normalize] / df[cols_to_normalize].max()

df[["length","width","height"]].head()
```

##  Binning Data (Discretization)

Converting continuous numerical data into discrete categories or bins to reduce noise and makes patterns more interpretable.

```python
#Since the "horsepower" is ranging from 48 to 288 and it has 59 unique values. 
#Since we only care about the price difference between cars, we can rearrange them into three ‘bins' high horsepower, medium horsepower, and little horsepower to simplify analysis.

import numpy as np

horsepower_bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ["Low", "Meduim", "High"]
df["horsepower-binned"] = pd.cut(df["horsepower"], horsepower_bins, labels=group_names, include_lowest=True)

df[['horsepower', 'horsepower-binned']].head(7)
```

```python
#visualizing binned data

import matplotlib.pyplot as plt

plt.bar(group_names, df["horsepower-binned"].value_counts())

#set x & y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show() 
```

It is clear that most cars have a low horsepower and only very few cars have high horsepower.

## Develop a model to predict car price: Model Evaluation and Refinement

### One-hot Encoding for Nominal Categorical Variables¶

```python
df = pd.get_dummies(df, columns=['fuel-system', "num-of-cylinders", "engine-type", "engine-location", "drive-wheels", "body-style", "num-of-doors", "aspiration", "fuel-type", "make"], drop_first=True)
```

### Label Encoding for Ordinal Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['horsepower-encoded'] = encoder.fit_transform(df['horsepower-binned'])

df["horsepower-binned"] = df["horsepower-binned"].astype(str)
df["horsepower-binned"] = encoder.fit_transform(df["horsepower-binned"])
```

### Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

X = df[['horsepower', 'curb-weight', 'engine-size', 'highway-L/100km', 'city-L/100km', 'width']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

print("number of test samples:", X_test.shape[0])
print("number of training samples:", X_train.shape[0])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['price'])
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

print("number of test samples:", X_test.shape[0])
print("number of training samples:", X_train.shape[0])
```

### Preprocess the Data

Since regression models work better with scaled data, let's standardize (normalize) the features

```python
#Check for non-numeric data
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, str)).sum() > 0:
        print(f"Column '{col}' has string values!")

#Convert all columns to numeric
df = df.apply(pd.to_numeric, errors='coerce')

#Handle NaNs and infinite values
df.fillna(0, inplace=True)
df.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Train and Evaluate a Machine Learning model

#### Model 1: Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Train a basic Linear Regression model
lm = LinearRegression()
lm.fit(X_train_scaled, y_train)
y_lr_pred = lm.predict(X_test_scaled)

print("Linear Regression R² Score:", r2_score(y_test, y_lr_pred))
print("Linear Regression MSE:", mean_squared_error(y_test, y_lr_pred)) 
```

#### Cross-validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lm, X_train_scaled, y_train, cv=5, scoring='r2')  # 5-fold cross-validation

print("Cross-validated R² scores:", scores)
print("Mean R² score:", scores.mean())
```

###### Since the R² score slightly dropped from 0.845 to 0.805 after cross-validation, then we will try for another model.

#### Model 2: Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=1, include_bias=False)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

#Fit the polynomial regression model
lm2 = LinearRegression()
lm2.fit(X_train_poly, y_train)

#Predict values
y_pr_pred = lm2.predict(X_test_poly) 

print("Polynomial Regression R² Score:", r2_score(y_test, y_pr_pred))
print("Polynomial Regression MSE:", mean_squared_error(y_test, y_pr_pred)) 
```

#### Cross-validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lm2, X_train_scaled, y_train, cv=5, scoring='r2')  # 5-fold cross-validation

print("Cross-validated R² scores:", scores)
print("Mean R² score:", scores.mean())
```

###### Since the R² score slightly dropped from 0.845 to 0.805 after cross-validation, then we will try for another model.

#### Model 3: Ridge Regression with Hyperparameter Tuning

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge_params = {"alpha": np.logspace(-3, 3, 10)}
ridge_grid = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='r2')
ridge_grid.fit(X_train_scaled, y_train)

best_ridge = ridge_grid.best_estimator_
y_rr_pred = best_ridge.predict(X_test_scaled)

print("Best Alpha:", ridge_grid.best_params_)    #Print the best alpha value

print("Ridge Regression R² Score:", r2_score(y_test, y_rr_pred))
print("Ridge Regression MSE:", mean_squared_error(y_test, y_rr_pred)) 
```

#### Cross-Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(ridge_grid, X_train_scaled, y_train, cv=5, scoring='r2')  # 5-fold cross-validation

print("Cross-validated R² scores:", scores)
print("Mean R² score:", scores.mean())
```

###### Since the R² score slightly dropped from 0.88 to 0.87 after cross-validation, then we will try for another model.

#### Model 4: Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=23)
rf_model.fit(X_train_scaled, y_train)
y_rf_pred = rf_model.predict(X_test_scaled)

print("Random Forest R² Score:", r2_score(y_test, y_rf_pred))
print("Random Forest MSE:", mean_squared_error(y_test, y_rf_pred)) 
```

#### Cross-Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf_model, X_train_scaled, y_train, cv=5, scoring='r2')  # 5-fold cross-validation

print("Cross-validated R² scores:", scores)
print("Mean R² score:", scores.mean())
```

###### Since the R² score slightly dropped from 0.885 to 0.90 after cross-validation, then we will consider this model.

#### Random Forest Model Evaluation Using Visualization

```python
#Kernel density estimate (KDE) plot

plt.figure(figsize=(10, 5))

#the actual price values
ax1 = sns.kdeplot(y_test, color="r", label="Actual Values")

#the predicted price values
sns.kdeplot(y_rf_pred, color="b", label="Predicted Values", ax=ax1)

ax1.set_xlabel("Price")
ax1.set_ylabel("Proportion of Cars")
ax1.set_title("Actual vs. Predicted Price Distribution")
ax1.legend()
plt.show()
```

We can see that the fitted values are reasonably close to the actual values since the two distributions overlap. However, the model tends to zero right below 30,000 dollars cars.

#### RandomForest Feature Importance

```python
# Get feature importance
importance = rf_model.feature_importances_

# Convert to DataFrame for better visualization
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importance})
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

# Plot Feature Importance
plt.figure(figsize=(20, 20))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='teal')
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance from RandomForest")
plt.gca().invert_yaxis()
plt.show()
```
